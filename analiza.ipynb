V tej nalogi bom analiziral statistiko prenosnih računalnikov na spletni trgovini [BigBang](https://www.bigbang.si/izdelki/racunalnistvo/prenosni-racunalniki/).

```python
import pandas as pd
import matplotlib.pyplot as plt

prenosniki = pd.read_csv('podatki/bigbang.csv', index_col='cena')

pd.options.display.min_rows = 20
```

Pogledjmo si najprej, kako je s številom prenosnikov glede na znamko.

```python
data = pd.read_csv('podatki/bigbang.csv', index_col='ime')

data_firma = data[data['firma'] != "Obnovljeno"]

stevila_prenosnikov = data_firma.groupby('firma')['firma'].count()

stevila_prenosnikov.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Znamka')
plt.ylabel('Število prenosnikov')
plt.title('Število prenosnikov glede na znamko')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

Kot pričakovano, imajo popularne znamke, kot so Apple, ASUS, HP in Lenovo naprodaj največ prenosnikov.

Kako je pa s ceno? Logično bi bilo, da bi bili gaming in multimedijski prenosniki najdražji. Zanima me tudi povprečna cena glede na znamko prenosnika.

```python
data = pd.read_csv('podatki/bigbang.csv', index_col='ime')

data['cena'] = data['cena'].str.replace('€', '').str.replace(',', '').str.replace('.', '').astype(float) / 100

data_firma = data[data['firma'] != "Obnovljeno"]
data_kategorija = data[data['kategorija'] != "Prenosni računalniki"]

povprecne_cene_firma = data_firma.groupby('firma')['cena'].mean()
povprecne_cene_kategorija = data_kategorija.groupby('kategorija')['cena'].mean()

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

povprecne_cene_kategorija.plot(kind='bar', color='orange', edgecolor='black', ax=axes[0])
axes[0].set_xlabel('Kategorija')
axes[0].set_ylabel('Povprečna cena (€)')
axes[0].set_title('Povprečna cena prenosnikov glede na kategorijo')

povprecne_cene_firma.plot(kind='bar', color='skyblue', edgecolor='black', ax=axes[1])
axes[1].set_xlabel('Znamka')
axes[1].set_ylabel('Povprečna cena (€)')
axes[1].set_title('Povprečna cena prenosnikov glede na znamko')

plt.tight_layout()

plt.show()
```

Res, multimedijski in gaming prenosniki so najdražji. Glede na firmo pa ima daleč najvišjo ceno Razer Blade, kar se sklada z grafom kategorij, ker v veliki večini prodaja gaming prenosnike. Med najcenejšimi pa so Chuwi, Dynabook, Fujitsu in Onda.

Zanimalo me je še, če obstaja kakšna povezava med popustom in ceno. Poglejmo si še grafični prikaz:

```python
data['popust_procenti'] = data['popust_procenti'].str.rstrip(' %').astype(float)

plt.scatter(data['cena'], data['popust_procenti'], color='skyblue', edgecolors='black')

plt.xlabel('Cena (€)')
plt.ylabel('Popust (%)')
plt.title('Popust v odvisnosti od cene prenosnikov')

plt.tight_layout()
plt.show()
```

Opazimo, da v povprečju odstotek popusta glede na ceno prenosnika pada in da, kar je mene presenetilo, sploh noben prenosnik s ceno čez 2000€ nima znižane cene. Zanimalo me je, zakaj, in to so razlogi, ki sem jih našel:
* omejena dobava,
* za razliko od cenenih prenosnikov, pri katerih je cena ponavadi veliko večja od cene posameznih delov, si proizvajalci dražjih ne morejo privoščiti znižati cene, ker so ponavadi narejeni iz dejansko dragih materialov, ki upravičujejo ceno prenosnika,
* večina kupcev, ki kupujejo visokokakovostne izdelke, si ne bo spremenila mnenja zaradi manjše razlike v ceni.

Za na konec bi bilo zanimivo videti še 10 najdražjih in 10 najcenejših prenosnih računalnikov. Zato jih razvrstimo po ceni:

```python

data_sortirano = data.sort_values(by='cena', ascending=False)

data_sortirano.drop(columns=['kategorija', 'popust_euro', 'prodajalec', 'popust_procenti'], inplace=True)

print(data_sortirano)
```
